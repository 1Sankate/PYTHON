```python
from dputils.scrape import Tag, Scraper   #tag class help to decide what to collect from the page # scraper executes the scraping process
url="https://www.amazon.in/s?k=laptop&crid=2RY4IPLSJE3GP&sprefix=la%2Caps%2C458&ref=nb_sb_ss_ts-doa-p_1_2"
amazon=Scraper(url)
target=Tag(cls='s-main-slot s-result-list s-search-results sg-row')
items=Tag(cls='s-result-item')
title=Tag('h2',cls='a-size-mini a-spacing-none a-color-base s-line-clamp-2')
price=Tag('span',cls='a-price-whole')
results=amazon.get_repeating_page_data(target,items,heading=title,price=price)
import pandas as pd
df=pd.DataFrame(results)
df
```

CLEANING CODEü§ç

```python
import numpy as np
def clean_price(price):
    if isinstance(price,str):
        price=price.replace('‚Çπ','').replace(',','').replace('.','').replace('e','')
        price=price.strip()
        if price.isnumeric():
            price=float(price)
            return price
        if len(price)==0:
            return np.nan
        return price
```

```python
clean_price('‚Çπ31,999')
```

```python
df['new_price']=df['price'].apply(clean_price)
df
```
