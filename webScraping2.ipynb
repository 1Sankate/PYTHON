```python
from dputils.scrape import Tag,Scraper
url='https://timesofindia.indiatimes.com/news'
news=Scraper(url)
target=Tag('ul',cls='HytnJ')
item=Tag('li')
title=Tag('p',cls='CRKrj')
description=Tag('p',cls='W4Hjm')
result=news.get_repeating_page_data(target,item,heading=title,description=description)
```

```python
import pandas as pd
import numpy as np
df=pd.DataFrame(result)
df
```
